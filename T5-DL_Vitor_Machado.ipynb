Imports e análise dos dados

```python
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Conv1D
```

```python
test = pd.read_csv('/kaggle/input/titanic/test.csv')
train = pd.read_csv('/kaggle/input/titanic/train.csv')
train.head()
```

```python
y_train = train['Survived']
y_train
```

```python
train.info()
```

Escolha das labels a serem usadas no treinamento

```python
labels = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']
x_train = train[labels]
x_train.head()
```

```python
x_train['Age'].fillna(x_train['Age'].mean(), inplace=True)
x_train.isna().sum()
```

```python
x_train['Pclass'] = x_train['Pclass']/1.
x_train['SibSp'] = x_train['SibSp']/1.
x_train['Parch'] = x_train['Parch']/1.
x_train = pd.get_dummies(x_train)
x_train.head()
```

```python
x_train['Sex_male'] = x_train['Sex_male']/1.
x_train['Sex_female'] = x_train['Sex_female']/1.
x_train.info()
```

```python
conv_model = Sequential([
    Conv1D(32, kernel_size=1, input_shape=(None, 6)),
    Dense(16,activation='sigmoid'),
    Conv1D(8, kernel_size=1),
    Dense(2, activation='sigmoid')
])
```

```python
conv_model.compile(loss='binary_crossentropy',
                  optimizer='adam')
```

```python
hist = conv_model.fit(x_train, y_train, epochs=50)
```

```python
model = Sequential([
  Dense(1000, activation='sigmoid', input_shape=(6,)),
  Dropout(0.3),
  Dense(500, activation='sigmoid'),
  Dropout(0.3),
  Dense(2, activation='softmax'),
])

model.compile(optimizer=Adam(),
              loss='binary_crossentropy',
              metrics=['accuracy'])

historico = model.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2
)
```

```python
loss = historico.history['loss']
accuracy = historico.history['accuracy']
val_loss = historico.history['val_loss']
val_accuracy = historico.history['val_accuracy']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
plt.xlim(10,50)
plt.ylim(0.69,0.7)
plt.plot(loss)
plt.plot(val_loss)
plt.legend(['Treino', 'Validação'])
```

```python
plt.plot(accuracy)
plt.plot(val_accuracy)
plt.legend(['Treino', 'Validação'])
```
